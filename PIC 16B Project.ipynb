```python
import pandas as pd
from sklearn.linear_model import LinearRegression

from plotly import express as px
```

```python
homevalue = pd.read_csv("Zip_home.csv")


homevalue.shape



homevalue = homevalue[homevalue["State"] == "CA"]

homevalue.rename(columns = {"RegionName":"Zip"}, inplace=True)

#Check Column Names and their indexes
#i = 0
#for col_name in homevalue.columns: 
   # print(col_name,i)
    # i = i+1

    
Cali_homevalue_subset1 = homevalue.iloc[:,177:312]
Cali_homevalue_subset2 = homevalue.iloc[:,[2,5,6]]

Cali_homevalue = pd.concat([Cali_homevaue_subset2, Cali_homevaue_subset1], axis=1)

Cali_homevalue = Cali_homevalue.set_index(["Zip","State","City"])

Cali_homevalue = Cali_homevalue.stack()



Cali_homevalue = Cali_homevalue.reset_index()
Cali_homevalue = Cali_homevalue.rename(columns = {"level_3" :"Date",0:"Value"})

Cali_homevalue["Year"] = Cali_homevalue["Date"].str[:4].astype(int)




Cali_homevalue_combinedMonth = Cali_homevalue.groupby(["City","Zip","Year"])["Value"].mean()



Cali_homevalue_combinedMonth = Cali_homevalue_combinedMonth.reset_index()
Cali_homevalue_combinedMonth





def coef(data_group):
    x = data_group[["Year"]] # 2 brackets because X should be a df
    y = data_group["Value"]   # 1 bracket because y should be a series
    LR = LinearRegression()
    LR.fit(x, y)
    return LR.coef_[0]




coefs = Cali_homevalue_combinedMonth.groupby(["City","Zip"]).apply(coef)
coefs = coefs.reset_index()

coefs 
```

```python
# California High School Data

Cali_school = pd.read_excel("California_highschool_ranking.xlsx")
Cali_school = Cali_school.iloc[3:,:]





Cali_school = Cali_school.rename(columns = Cali_school.iloc[0,:])

Cali_school = Cali_school.iloc[1:,:]
Cali_school = Cali_school.reset_index()
Cali_school = Cali_school.drop("index", axis = 1)
Cali_school["Zip"] = Cali_school["Zip"].astype(int)


# Access to top schools
#ToptenSchools = Cali_school.iloc[:300,:]

#ToptenSchools["Percent Asian"] = ToptenSchools["Percent Asian"].str[:-1].astype(float)

#ToptenSchools["Percent Asian"].mean()



zip_Geoinfo = pd.read_csv("us-zip-code-latitude-and-longitude.csv",sep=";") 

zip_Geoinfo["Zip"] = zip_Geoinfo["Zip"].astype(int)



Cali_School_Geoinfo = Cali_school.merge(zip_Geoinfo, on = "Zip")

Cali_School_Geoinfo["Latitude"] = Cali_School_Geoinfo["Latitude"].astype(float)
Cali_School_Geoinfo["Longitude"] = Cali_School_Geoinfo["Longitude"].astype(float)
Cali_School_Geoinfo["Rank (2018-19)"] = Cali_School_Geoinfo["Rank (2018-19)"].astype(float)




fig = px.scatter_mapbox(Cali_School_Geoinfo,
                            lat = "Latitude",
                            lon = "Longitude",
                            hover_name = "School",
                            color = "Rank (2018-19)",
                            zoom = 1,
                            height = 300,
                            mapbox_style = "stamen-terrain",
                            labels = {"ABC"},
                            title = "a")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})

fig.show()
```
